```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sympy import false
```

```python
df_azure_cold = pd.read_csv("./data/azure-cold-start.csv")
df_azure_warm = pd.read_csv("./data/azure-warm-start.csv")
df_openfaas_cold = pd.read_csv("./data/openfaas-cold-start.csv")
df_openfaas_warm = pd.read_csv("./data/openfaas-warm-start.csv")
```

```python
# 数据预处理
df_azure_cold = df_azure_cold[df_azure_cold['label']=='HTTP Request-1']
df_azure_warm = df_azure_warm[df_azure_warm['label']=='HTTP Request-1']

df_azure_cold_latency = df_azure_cold['Latency']
df_azure_warm_latency = df_azure_warm['Latency']
df_openfaas_cold_latency = df_openfaas_cold['Latency']
df_openfaas_warm_latency = df_openfaas_warm['Latency']
```

```python

plt.figure(figsize=(5, 5))
plt.boxplot([df_azure_cold_latency,df_openfaas_cold_latency,],vert=False, tick_labels=["azure_cold_latency","openfaas_cold_latency"])
plt.title("cold start")
plt.xlabel("Latency/ms")
plt.show()
```

```python

plt.figure(figsize=(5, 5))
plt.boxplot([df_azure_warm_latency,df_openfaas_warm_latency,],vert=False, tick_labels=["azure_warm_latency","openfaas_warm_latency"])
plt.title("warm start")
plt.xlabel("Latency/ms")
plt.show()
```

```python
# create subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))  # 1行2列的子图

axs[0].boxplot([df_azure_cold_latency,df_openfaas_cold_latency,],vert=False, tick_labels=["azure_cold_latency","openfaas_cold_latency"])
axs[0].set_title("cold start")
axs[0].set_xlabel("Latency/ms")

axs[1].boxplot([df_azure_warm_latency,df_openfaas_warm_latency,],vert=False, tick_labels=["azure_warm_latency","openfaas_warm_latency"])
axs[1].set_title("warm start")
axs[1].set_xlabel("Latency/ms")

# 调整间距并显示
plt.tight_layout()
plt.show()
```
